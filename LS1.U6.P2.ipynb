```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(-3, 4, .01)
A = 1
y1 = A * np.cos(7 * np.pi * t)
y2 = A * np.cos(9 * np.pi * t)
y3 = y1 + y2

step_x = np.arange(-3, 4, .5)
step_y = np.arange(-2, 3, .5)
plt.grid(True)
plt.xticks(step_x)
plt.yticks(step_y)
plt.plot(t, y3)

plt.show()
```
