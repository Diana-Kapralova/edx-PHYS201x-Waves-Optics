```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 18, 1)
A = np.array([10, 8, 6, 5, 4.5, 4, 3.8, 3.2, 3, 2.8, 2.7, 2.5, 2.2, 2.1, 2, 1.9, 1.7, 1.6])
z = np.polyfit(t, A, 3)
p = np.poly1d(z)
A_tmp = p(t)
plt.scatter(t, A)
plt.plot(t, A_tmp)

plt.ylim([0, 10])
plt.xlim([0, 20])
plt.xlabel('time (min)')
plt.ylabel('amplitude (mm)')
plt.title('Dambed SHM')
plt.grid(True)
plt.show()
```
