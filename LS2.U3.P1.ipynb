```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-4, 5, 1)
L = 4
A0 = 0
A1 = -0.08 * L
A2 = 0
A3 = 0.02 * L
B1 = 0
B2 = -0.1 * L
yA = A0/2 + A1 * np.cos(np.pi * x/L) + A2 * np.cos( 2 * np.pi * x/L) + A3 * np.cos(3 * np.pi * x/L)
yB = B1 * np.sin(np.pi * x/L) + B2 * np.sin( 2 * np.pi * x/L)
y = yA + yB


dots_fx = x
dots_fy = np.array([0, 0, 0, 0, 0, -1, 0, 1, 0])

plt.grid(True)
plt.plot(x, y)
plt.plot(dots_fx, dots_fy, "C1")

plt.show()
```
